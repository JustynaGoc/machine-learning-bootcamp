```python
import numpy as np
```

```python
X1 = np.array([1,2,3,4,5,6])
Y = np.array([3000,3250,3500,3750,4000,4250])
X1,Y
```

```python
m = len(X1)
print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba probek: {m}')
```

```python
X1=X1.reshape(m,1)
X1
```

```python
bias = np.ones((m,1))
bias
```

```python
X = np.append(bias,X1,axis=1)
X
```

```python
np.dot(X.T, X)
```

```python
L = np.linalg.inv(np.dot(X.T, X))
L
```

```python
P=np.dot(X.T,Y)
P
```

```python
np.dot(L,P)
```
