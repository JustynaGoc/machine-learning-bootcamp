#Przygotowanie danych do uczenie maszynowego - data processuing

```python
##Importowanie bibliotek
```

```python
import pandas as pd
import numpy as np
import sklearn
sklearn.__version__
```

```python
##Tworzenie zbioru do ćwiczeń
```

```python
data = {
    "size" :  ["XL", "L", "M", "L", "M"],
    "color":  ["red", "green", "blue", "green", "red"],
    "gender": ["female", "male", "male", "female", "female"],
    "price":  [199.0, 89.0, 99.0, 129.0,  79.0],
    "weight": [500, 450, 300, 380, 410],
    "bought": ["yes", "no", "yes", "no", "yes"]
}
df_raw = pd.DataFrame(data)
df_raw
```

```python
##utworzenie kopii zapasowej
```

```python
df = df_raw.copy()
df.info()
```

```python
##zmiana typy danych  i wstepna eksploracja
```

```python
for col in ["size", "color", "gender", "bought"]:
  df[col] = df[col].astype('category')

df["weight"]= df["weight"].astype(float)
df.info()
```

```python
df.describe().T
#domyślnie  tylko  dla kolumn liczbowych
# .T  transponuje dane na bardziej czytelny  format
```

```python
df.describe(include =["category"]).T
#statystyki dala  kolumn kategorycznych
```

```python
##std() = nandas nieobciążony- dzieli przez N-1
##std() - numpy obciążony - dzieli przez N
```

```python
print(f"{df["price"]}\n")
print(f"Średinia pandas : {df["price"].mean()}")
print(f"Odchylenie standardowe pandas : {df["price"].std()}")

print(f"Średinia numpy: {np.mean(df["price"])}")
print(f"Odchylenie standardowej:numpy {np.std(df["price"])}")
```

```python
(df['price'] -df['price'].mean()) /df['price'].std()
```

```python
def standardize(x):
  return(x - x.mean())/x.std()
standardize(df["price"])
```

```python
def standardize(x):
  return(x - np.mean(x))/np.std(x)
standardize(df["price"])
```
