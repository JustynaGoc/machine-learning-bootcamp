```python
def fetch_financial_data(company = "AMZN"):
  """Ta funkcja pobiera notowania gie≈Çdowe"""
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source= 'stooq')

df_raw =fetch_financial_data()
df_raw.head()
```

```python
df_raw.info()
```

```python
df= df_raw.copy()
df=df[:5]
df.info()
```

```python
df["year"]= df.index.year
df["month"]= df.index.month
df["day"]= df.index.day
df["day_name"]= df.index.day_name()
df
```

```python
df = pd.DataFrame(data={'height': [160.,175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
df["height_cat"] = pd.cut(x=df.height, bins= 3)
df
```

```python
df["height_cat"]= pd.cut(df.height, bins = (159,175,180,195), labels = ["small", "medium", "high"])
df
```

```python
pd.get_dummies(df, drop_first = True, dtype= int)
```

```python
df = pd.DataFrame(data = {"lang": [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python
df["lang_number"]= df["lang"].apply(len)
df
```

```python
df["pl_flag"]= df["lang"].apply(lambda x:1 if "PL" in x else 0)
df
```
