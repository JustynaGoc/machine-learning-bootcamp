```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.datasets import make_regression
```

```python
sns.set(font_scale=1.3)
np.random.seed(42)
np.set_printoptions(suppress=True, precision=6)
sklearn.__version__
```

```python
#wygenerowanie danych
```

```python
data, target  = make_regression(n_samples = 200, n_features=1, noise =20 )
target = target**2
```

```python
print(data[:5]), print(target[:5])
```

```python
#Regresja liniowa
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data, target, label = "dane")
plt.title("Regresja liniowa")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
from  sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(data, target)
```

```python
x_min = np.floor(data.min())
x_max = np.ceil(data.max())
x_min,x_max

plot_data = np.arange(-3, 3, 0.01).reshape(-1, 1)

plt.figure(figsize =(8,6))
plt.title("Regresja Liniowa")
plt.xlabel("cecha x")
plt.ylabel("zmienna docelowa")
plt.scatter(data, target, label = "cecha x")
plt.plot(plot_data, regressor.predict(plot_data), color = 'black', label = 'Regresja liniowa')
plt.legend()
plt.show()
```

```python
x_min = np.floor(data.min())
x_max = np.ceil(data.max())
x_min, x_max
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor  = DecisionTreeRegressor()
regressor.fit(data, target)
```

```python
plt.figure(figsize =(8,6))
plt.title("Regresja drzewa decyzyjneg")
plt.xlabel("cecha x")
plt.ylabel("zmienna docelowa")
plt.scatter(data, target, label = "cecha x")
plt.plot(plot_data, regressor.predict(plot_data), color = 'black', label = 'Regresja drzewa decyzyjnego')
plt.legend()
plt.show()
```

```python
# glebokosc drzewa decyzyjnego = 1
max_dept = 1
regressor  = DecisionTreeRegressor(max_depth=max_dept)
regressor.fit(data, target)
plt.figure(figsize =(8,6))
plt.title("Regresja drzewa decyzyjnegO - glebokosc drzewa decyzyjnego -1")
plt.xlabel("cecha x")
plt.ylabel("zmienna docelowa")
plt.scatter(data, target, label = "cecha x")
plt.plot(plot_data, regressor.predict(plot_data), color = 'black', label = 'Regresja drzewa decyzyjnego - glebokosc 1')
plt.legend()
plt.show()
```

```python
# glebokosc drzewa decyzyjnego = 2
max_dept = 2
regressor  = DecisionTreeRegressor(max_depth=max_dept)
regressor.fit(data, target)
plt.figure(figsize =(8,6))
plt.title("Regresja drzewa decyzyjnegO - glebokosc drzewa decyzyjnego -2")
plt.xlabel("cecha x")
plt.ylabel("zmienna docelowa")
plt.scatter(data, target, label = "cecha x")
plt.plot(plot_data, regressor.predict(plot_data), color = 'black', label = 'Regresja drzewa decyzyjnego - glebokosc 2')
plt.legend()
plt.show()
```

```python
from os import write
from sklearn.tree import export_graphviz
from io import StringIO
from IPython.display import Image
import pydotplus
dot_data = StringIO()
export_graphviz(regressor, out_file = dot_data, filled = True, rounded = True, special_characters = True, feature_names = ["cecha x"])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('drzewo_decyzyjne.png')
Image(graph.create_png(), width = 900)
```

```python
def make_dt_regression(max_dept = 2):
  regressor  = DecisionTreeRegressor(max_depth=max_dept)
  regressor.fit(data, target)
  plot_data = np.arange(np.floor(data.min()), np.ceil(data.max()), 0.01).reshape(-1, 1)

  plt.figure(figsize =(8,6))
  plt.scatter(data, target)
  plt.plot(plot_data, regressor.predict(plot_data), c = 'red')

  dot_data = StringIO()
  export_graphviz(regressor, out_file = dot_data, filled = True, rounded = True, special_characters = True, feature_names = ["cecha x"])
  graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
  graph.write_png('drzewo_decyzyjne.png')
  return Image(graph.create_png(), width = 300 + max_dept * 100)
make_dt_regression(max_dept = 2)
```
