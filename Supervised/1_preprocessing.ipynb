# Przygotowanie danych

##1. Import bibliotek

```python
import pandas as pd
import numpy as np
import sklearn
```

```python
sklearn.__version__
```

## 2. Wygenerowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', '  female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data = data)
df_raw
```

## 3. Utworzenie kopii danych

```python
df = df_raw.copy()
```

```python
df.info()
```

## 4. Zmiana typu danych

```python
#zmiana typy danych object na kategoryczne - dobre praktyki

for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight']  = df['weight'].astype('float')
```

```python
df.info()
```

```python
# automatyzacja
df = df_raw.copy()
df.info()
```

```python
for col in  df.select_dtypes(include=['object']).columns.tolist():
  df[col] = df[col].astype('category')
for col in  df.select_dtypes(include=['int64']).columns.tolist():
  df[col] = df[col].astype('float')
df.info()
```

```python
df.describe().T
# domyślnie   zwraca  statystyki wylacznie danych numerycznych
```

```python
df.describe(include = ['category']).T
```

## 6. OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output = False, drop = 'first')
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
df
```

```python
df_add = pd.DataFrame(encoder.fit_transform(df[['size']]), columns = encoder.get_feature_names_out(['size']))
df_add
```
