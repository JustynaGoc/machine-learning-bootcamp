```python
import pandas as pd
```

```python
data = {'size': ["XL", "L", "M", "L", "M"],
        'color':["red", "green", "blue", "green", "red"],
        'gender':["female", "male", "male", "female", "female"],
        'price':[199.0, 89.0, 99.0, 129.0, 79.0],
        'weight':[500, 450, 300, 380, 410],
        'bought':["yes", "no", "yes", "no", "yes"]
        }
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

zamiana zmiennej  będącej targetem na postać  0/1 --- zmienna bought --- nasz  cel

```python
mapping = {'yes': 1, 'no': 0}
df['bought2'] = df['bought'].map(mapping)
df
```

```python
df.drop('bought2', axis=1, inplace=True)
df
```

```python
pd.get_dummies(df)
df
```

```python
mean = df['price'].mean()
std = df['price'].std()
df['price_norm'] = (df['price'] - mean)/std
df
mean = df['weight'].mean()
std = df['weight'].std()
df['weight_norm'] = (df['price'] - mean)/std
df

```
